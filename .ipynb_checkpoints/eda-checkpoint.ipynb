```python
import pandas as pd

from nltk.corpus import stopwords, wordnet
from nltk.probability import FreqDist
from nltk.stem import WordNetLemmatizer
from nltk.tag import pos_tag
from nltk.tokenize import TweetTokenizer

import matplotlib.pyplot as plt 
import seaborn as sns 

import re
```

### 1. Load the dataset

The dataset is originally taken from a Kaggle competition. We use pandas for ease of manipulation of the data. 

```python
# load dataset 
url = "https://raw.github.com/ant1code/tweet-sentiment/master/data/train.csv"
df = pd.DataFrame(data=pd.read_csv(url), columns=["text", "sentiment"])

# quick preview of dataset 
df.head()
```

### 2. Prepare the dataset

Note that the categorical sentiment variable is not suitable for machine learning and needs to be converted to a numerical label. Rows with missing values should also be dropped because we are performing supervised learning. 

```python
# convert sentiment (categorical) to label (numerical)
# negative = 0, neutral = 1, positive = 2
df["sentiment"]  = df.sentiment.map({"negative": 0, "neutral": 1, "positive": 2}) 

# remove missing values 
df = df[df.text.notna()]

# peek at the dataset again
df.head()
```

### 3. Preliminary exploration of the dataset 

We can now do some simple exploration of the dataset, such as looking at the number of tweets, the average number of words per sentiment, and the distribution of the sentiments. The second task is interesting because it allows us to have an idea of whether or not the number of words in a tweet could be correlated with the sentiment it has. However, the third task is the most important because we will be conducting supervised learning later on, and an unbalanced dataset could lead to a biased classifier. 

```python
# get length of longest sentence in entire dataset
max_length = max(df.text.apply(len))

print(f"There are {len(df)} tweets in this dataset.")
print(f"The longest tweet in this dataset has {max_length} words.")
```

```python
# get average number of words per sentiment 
df["word_count"] = df.text.apply(lambda x: len(str(x).split()))
avg_word_count = df.groupby("sentiment")["word_count"].mean()

sentiments = ["negative", "neutral", "positive"]

plt.style.use(["seaborn"])
plt.figure(figsize=(50,30))
sns.barplot(sentiments, avg_word_count, alpha=0.8)

plt.title("Average Number of Words by Sentiment", fontsize=80)
plt.xlabel("Sentiment", fontsize=50)
plt.xticks(fontsize=50)
plt.ylabel("Average Number of Words", fontsize=50)
plt.yticks(fontsize=50)

plt.show()
```

### 3. Normalize the dataset

Fortunately for us, the dataset appears to be (at least roughly) balanced. The next step is to *normalize* the dataset. We explain what this process entails with the following made-up tweet: 
- "I'm in LOVE with this game! It makes me feel so relaxed and happy @nintendo"

This tweet has a positive sentiment. However, only some words (namely "LOVE", "relaxed", "happy") were helpful in getting to this conclusion; the other words are either fairly neutral (e.g. "game", "makes") or are common parts of everyday speech (e.g. "in", "with", "and"). This latter group of words is called *stop words*. 

Since the complexity of text-processing models generally scale with the number of words involved, it makes sense to remove words that do not contribute significantly to the sentiment of the tweet as far as possible. The NLTK package contains a list of common stop words, and after adapting this list to suit our project, we can simply filter out these stop words. The most important change we make is to remove words like "no", "not", and "couldn't" from the list, since these clearly are negative and would have an impact on the overall tweet sentiment. 

```python
# get list of standard stop words
stop_words = stopwords.words("english")

# remove negations from stop_words (no, nor, not)
del stop_words[116:119]

# remove negative contractions from stop_words (e.g. couldn't)
del stop_words[128:130]
del stop_words[123:125]
del stop_words[129:]

# add neutral contractions to stop_words (e.g. they'll)
stop_words.extend([
    "i'm", "i'll", "i'd", "us", "it'll", "it'd", 
    "they're", "they'd", "they'll", 
    "whose", "who're", "who\'ll", "who'd", 
    "that's"
])

# peek at list of stop words
print(stop_words[:10])
```

```python
s = "I'm in LOVE with this game! It makes me feel so relaxed and happy @nintendo" 

# NLTK's stopwords are all in lowercase, so we need to lowercase s 
s = s.lower()

# filtering out stop words from the made-up tweet 
no_stop_words = " ".join([w for w in s.split(" ") if w not in stop_words])
print(no_stop_words)
```

The next most important task is called *lemmatizing*, which is the act of reducing words to their root forms. The main reasons behind lemmatizing are: 
- cutting down on the range of words used in the dataset and thereby speeding up the training process;
- the inflectional endings of most words do not have a large impact on the effect they have on the sentiment of the sentence. For instance, the word "laughs" and its root word "laugh" are both positive. 

However, the way to lemmatize a word can be ambiguous because some words can be a noun or a verb depending on the context. For instance, the word "writing" could be either refer to the noun or the verb "write" in the present continuous tense. It is therefore crucial to assign each word its *part of speech* (POS) tag before lemmatizing the tweet. This is done using NLTK's `pos_tag`. 

```python
# renaming the POS tags to suit our needs
def get_wordnet_pos(tag):
    if tag.startswith("J"):
        return wordnet.ADJ 
    elif tag.startswith("V"):
        return wordnet.VERB
    elif tag.startswith("N"):
        return wordnet.NOUN
    elif tag.startswith("R"):
        return wordnet.ADV
    else:
        return "" 
    
# loading the lemmatizer 
lemmatizer = WordNetLemmatizer()
```

```python
# tagging each word in the filtered made-up tweet
tags = pos_tag(no_stop_words.split(" "))
tags = [(w, get_wordnet_pos(t))for w,t in tags]
print(tags)

# lemmatizing the filtered made-up tweet based on the tags 
lemmatized = [lemmatizer.lemmatize(w, pos) for w,pos in tags]
print(lemmatized)
```

Observe that although the words have been reduced to root form correctly, some of the words still have punctuation attached. These punctuation need to be separated and removed. Other miscellaneous items that should be removed include numbers, links, and Twitter handles. 

Below we define the function `normalize`, which does everything in this step for us. This function is then applied to our entire dataset. 

```python
def _normalize(text, stop_words, lemmatizer=WordNetLemmatizer()):
    text = str(text).lower()
    
    no_twitterIDs = re.sub(r"@\S+", "", text)
    no_urls = re.sub(re.compile(r"http\S+|www\S+"), "", no_twitterIDs)
    no_numbers = re.sub("[0-9]+", "", no_urls)
    
    tokenizer = TweetTokenizer()
    tokenized = tokenizer.tokenize(no_numbers)
    
    re_words = re.compile(r"\w+")
    no_punctuation = [w for w in tokenized if re_words.search(w) and w not in stop_words]
    
    tagged = pos_tag(no_punctuation)
    
    lemmatized = []
    for word, tag in tagged:
        pos = get_wordnet_pos(tag)
        if pos:
            lemma = lemmatizer.lemmatize(word, pos)
        else:
            lemma = word
        lemmatized.append(lemma)
    return lemmatized
```

```python
# normalising the made-up tweet
print(_normalize(s, stop_words, lemmatizer))
```

```python
# normalizing the dataset 
df["lemmatized"] = df.text.apply(lambda x: _normalize(x, stop_words))
df.lemmatized = df.lemmatized.apply(lambda x: " ".join(x))

# get length of longest sentence in entire dataset
max_length = max(df.lemmatized.apply(len))

print(f"The longest (normalized) tweet in this dataset has {max_length} words.")
df[["text", "lemmatized"]].head()
```

### 4. Exploration of the normalized dataset

In this section we explore the following hypotheses: 
- Is the number of words correlated with the tweet sentiment? 
- What are the most common words for each sentiment, and is the presence of these words a reliable indicator of the tweet sentiment? 

```python
# is number of words correlated with sentiment?
df["word_count"] = df.lemmatized.apply(lambda x: len(str(x).split()))
avg_word_count = df.groupby("sentiment")["word_count"].mean()

plt.figure(figsize=(50,30))
sns.barplot(sentiments, avg_word_count, alpha=0.8)

plt.title("Average Number of Words by Sentiment", fontsize=80)
plt.xlabel("Sentiment", fontsize=50)
plt.xticks(fontsize=50)
plt.ylabel("Average Number of Words", fontsize=50)
plt.yticks(fontsize=50)

plt.show()
```

The average number of words is roughly the same across the sentiments. This suggests that looking at the number of words in a tweet would be a poor predictor of the tweet's sentiment. 

Note: this graph is surprisingly similar to the first graph depicting the average number of words by sentiment in the original dataset. But it makes sense that the average number of words per tweet has decreased by the same amount across the sentiments, since the words removed were (mainly) stop words. 

```python
# What are the top k=20 words per sentiment? 
def get_topk_words(df, sentiment, k=20):
    tweets = df[df["sentiment"]==sentiment]
    
    all_words = []
    for t in tweets.lemmatized:
        all_words.extend(t.split(" "))
    
    topk_words = FreqDist(all_words).most_common(k)
    x, y = zip(*topk_words)

    plt.figure(figsize=(50,30))
    plt.bar(x, y)
    plt.title("Frequency of {} most common words".format(k), fontsize=80)
    plt.xlabel("Words", fontsize=50)
    plt.xticks(fontsize=50, rotation=60)
    plt.ylabel("Frequency", fontsize=50)
    plt.yticks(fontsize=50)

    plt.show()
    return dict(topk_words)
```

```python
# sentiment=0: negative sentiment 
topk_negative = get_topk_words(df, sentiment=0)
```

```python
# label=1: neutral sentiment 
topk_neutral = get_topk_words(df, sentiment=1)
```

```python
# label=2: positive sentiment 
topk_positive = get_topk_words(df, sentiment=2)
```

From the graphs above, one might conclude that the most frequent word per sentiment are good ways to characterise that sentiment. We explore this hypothesis by defining the function `freq_ratio`, which helps us measure how often a word appears in one sentiment compared to in other sentiments. 

```python
# return comparison of frequency of a word 
# < 1: more common in other sentiments
# = 1: equally likely to be found in this sentiment or the others
# > 1: more common in this sentiment
def freq_ratio(freq, freq_total):
    return round(freq/freq_total, 5) 
```

```python
# do the top 5 words in positive tweets occur frequently in negative tweets?
print("Checking occurrences of the most common positive words in negative tweets...")
top5_positive = list(topk_positive)[:5]
for word in top5_positive:
    freq = topk_positive[word]
    nonpositive_lemmatized = df[df["sentiment"] != 2].lemmatized
    freq_total = sum([lst.count(word) for lst in nonpositive_lemmatized])
    print(f"{word}: ", freq_ratio(freq, freq_total))
```

```python
# do the top 5 words in negative tweets occur frequently in positive tweets?
print("Checking occurrences of the most common negative words in positive tweets...")
top5_negative = list(topk_negative)[:5]
for word in top5_negative:
    freq1 = topk_negative[word]
    positive_lemmatized = df[df["sentiment"] == 1].lemmatized
    freq2 = sum([lst.count(word) for lst in positive_lemmatized])
    print(f"{word}: ", freq_ratio(freq1, freq2))
```

In both instances, the most common word achieves a value lower than 1 and so is not particularly characteristic of the sentiment. This suggests that we cannot simply take the most common word per sentiment and use it as a predictor. (Interestingly, words like "love" and "happy" achieve very high values > 1, yet none of them are the most common word among positive tweets.) 

This exploratory data analysis shows that there is a need for more advanced techniques when it comes to sentiment analysis of tweets. In the other modules, we implement a Random Forest classifier and fine-tuned a pretrained (Distil)Bert classifier, which prove to be signficantly more reliable predictors of tweet sentiment.
