```python
import pandas as pd

from nltk.corpus import stopwords, wordnet
from nltk.probability import FreqDist
from nltk.stem import WordNetLemmatizer
from nltk.tag import pos_tag
from nltk.tokenize import TweetTokenizer

import matplotlib.pyplot as plt 
import seaborn as sns 

import re
```

### 1. Load the dataset

The dataset is originally taken from a Kaggle competition. We use pandas for ease of manipulation of the data. 

```python
url = "https://raw.github.com/ant1code/tweet-sentiment/master/data/train.csv"
df = pd.read_csv(url)

# quick preview of dataset 
df.head()
```

### 2. Clean the dataset

Note that the categorical sentiment variable is not suitable for machine learning and needs to be converted to a numerical label. Additionally, the columns `textID` and `selected_text` are not relevant to this project and should be removed to reduce memory usage. Rows with missing values should also be dropped because we are performing supervised learning. 

```python
# convert sentiment (categorical) to label (numerical)
# negative = 0, neutral = 1, positive = 2
sentiment_to_label = {"negative": 0, "neutral": 1, "positive": 2}
df["label"]  = df.sentiment.map(sentiment_to_label) 

# drop irrelevant columns, remove missing values 
df.drop(["textID", "selected_text", "sentiment"], axis=1, inplace=True)
df = df[df.text.notna()]

# get length of longest sentence in entire dataset
max_length = max(df.text.apply(len))

print(df.head())
print(f"There are {len(df)} tweets in this dataset.")
print(f"The longest tweet in this dataset has {max_length} words.")
```

### 3. Preliminary exploration of the dataset 

We can now explore the dataset to understand it better. In particular it is important to know if the dataset is balanced, i.e. if the labels are evenly distributed, to avoid getting a biased sentiment classifier. 

```python
# *Preliminary exploration of the dataset*
# histogram of sentiments 
x = df.label.value_counts().sort_index()

plt.figure(figsize=(50,30))
plt.style.use(["seaborn"])
sns.barplot(list(sentiment_to_label), x.values, alpha=0.8)

plt.title("Sentiment Frequency", fontsize=80)
plt.xlabel("Sentiment", fontsize=50)
plt.xticks(fontsize=50)
plt.ylabel("Frequency", fontsize=50)
plt.yticks(fontsize=50)

plt.style.use(["seaborn"])
plt.show()

# pie chart of sentiments of dataset
plt.figure(figsize=(30,10))
x.plot(kind="pie", labels=list(sentiment_to_label), wedgeprops=dict(width=0.7), 
      autopct="%0.1f%%", startangle=90, textprops={"fontsize": 15})
plt.title("Distribution of Sentiments", fontsize=25)
plt.ylabel("")

plt.show()
```

### 3. Prepare for data exploration 

Before the dataset is ready for any more in-depth analysis, it has to be normalised: 
- all words should be converted to lowercase;
- non-words (e.g. twitterIDs, URLs, numbers, punctuation) should be removed; 
- words that do not contain great significance to the sentiment (e.g. pronouns, prepositions) should be removed; 
- words should be converted to their root forms wherever possible (e.g. eating -> eat) 

The first two subtasks can be accomplished easily using the `string` and `regex` packages. 

Words to be removed in the third subtask are known as *stop words* and a list of common stop words can be found in the `nltk` package. However, this list is not perfectly suited to our goal of sentiment analysis - negations (e.g. no, nor, not) clearly have a large influence on the sentiment of the tweet! Contractions (e.g. don't, musn't, it's) also need to be handled carefully because some contractions are also negative (e.g. don't). We manually remove contractions with implicit negative sentiments (and also add in some missing contractions). 

```python
# preparing the list of stop words
stop_words = stopwords.words("english")

# remove negations from stop_words (no, nor, not)
del stop_words[116:119]

# remove negative contractions from stop_words (e.g. couldn't)
del stop_words[128:130]
del stop_words[123:125]
del stop_words[129:]

# add neutral contractions to stop_words (e.g. they'll)
stop_words.extend([
    "i'm", "i'll", "i'd", "us", "it'll", "it'd", 
    "they're", "they'd", "they'll", 
    "whose", "who're", "who\'ll", "who'd", 
    "that's"
])
```

The fourth subtask is called *lemmatizing*. This is important because: 
- it cuts down on the range of words used in the dataset and therefore helps speed up the training process;
- the inflectional endings of most words do not have a large impact on the effect they have on the sentiment of the sentence. For instance, the word "laughing" and its root word "laugh" are both positive. 

However the way to lemmatize a word can be ambiguous because some words can be a noun or a verb depending on the context. Thus before lemmatizing, it is important to assign each word a tag representing its Part of Speech (POS). 

```python
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return ''
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess(text, stop_words, lemmatizer):
    text = str(text).lower()
    
    no_twitterIDs = re.sub(r"@\S+", "", text)
    no_urls = re.sub(r"http\S+", "", no_twitterIDs)
    no_numbers = re.sub("[0-9]+", "", no_urls)
    no_stop_words = " ".join([w for w in no_numbers.split(" ") if w not in stop_words])
    
    tokenizer = TweetTokenizer()
    tokenized = tokenizer.tokenize(no_stop_words)
    
    re_words = re.compile(r"\w+")
    no_punctuation = [w for w in tokenized if re_words.search(w)]
    
    tagged = pos_tag(no_punctuation)
    
    lemmatized = []
    for word, tag in tagged:
        pos = get_wordnet_pos(tag)
        if pos:
            lemma = lemmatizer.lemmatize(word, pos)
        else:
            lemma = word
        lemmatized.append(lemma)
    return lemmatized

# example sentence
s = "I don't. like. eating. big. oranges!!! This is 2020. Stop... @twitter https://www.twitter.com"

print("original sentence: ", s)
print("processed sentence: ", " ".join(preprocess(s, stop_words, lemmatizer)))
```

```python
# apply to dataset 
df["lemmatized"] = df.text.apply(lambda x: preprocess(x, stop_words, lemmatizer))
df.text = df.lemmatized.apply(lambda x: " ".join(x))

# get length of longest sentence in entire dataset
max_length = max(df.text.apply(len))

print(df.head())
print(f"The longest (processed) tweet in this dataset has {max_length} words.")
```

### 4. Exploration of the processed dataset

In this section we explore the following hypotheses: 
- Is the number of words correlated with the sentiment of the tweet? 
- Are certain words correlated with certain sentiments, and is the presence of such words in a tweet a reliable indicator of what sentiment the tweet has? 

```python
# is number of words correlated with sentiment?
df["word_count"] = df.text.apply(lambda x: len(str(x).split()))
avg_word_count = df.groupby("label")["word_count"].mean()

plt.figure(figsize=(50,30))
sns.barplot(list(sentiment_to_label), avg_word_count, alpha=0.8)

plt.title("Average Number of Words by Sentiment", fontsize=80)
plt.xlabel("Sentiment", fontsize=50)
plt.xticks(fontsize=50)
plt.ylabel("Average Number of Words", fontsize=50)
plt.yticks(fontsize=50)

plt.show()
```

The average number of words is more or less the same across the three sentiments. This suggests that looking at number of words alone would be a poor way to predict sentiment. 

```python
# What are the top k=20 words per sentiment? 
def get_topk_words(df, label, k=20):
    tweets = df[df["label"]==label]
    
    all_words = []
    for t in tweets.text:
        all_words.extend(t.split(" "))
    
    topk_words = FreqDist(all_words).most_common(k)
    x, y = zip(*topk_words)

    plt.figure(figsize=(50,30))
    plt.bar(x, y)
    plt.title("Frequency of {} most common words for label={}".format(k, label), fontsize=80)
    plt.xlabel("Words", fontsize=50)
    plt.xticks(fontsize=50, rotation=60)
    plt.ylabel("Frequency", fontsize=50)
    plt.yticks(fontsize=50)

    plt.show()
    return dict(topk_words)
```

```python
# label=0: negative sentiment 
topk_negative = get_topk_words(df, label=0)
```

```python
# label=1: neutral sentiment 
topk_neutral = get_topk_words(df, label=1)
```

```python
# label=2: positive sentiment 
topk_positive = get_topk_words(df, label=2)
```

```python
# is it meaningful for some words to be used to characterise the sentiment? 
def pos_rate(freq1, freq2):
    return round(freq1/(freq1 + freq2), 5)

# do the top 5 words in positive tweets occur frequently in negative tweets?
print("Checking occurrences of positive words in negative tweets...")
top5_positive = list(topk_positive)[:5]
for word in top5_positive:
    freq1 = topk_positive[word]
    negative_lemmatized = df[df["label"] == 0].lemmatized
    freq2 = sum([lst.count(word) for lst in negative_lemmatized])
    print(f"{word}: ", pos_rate(freq1, freq2))

# do the top 5 words in negative tweets occur frequently in positive tweets?
print("Checking occurrences of negative words in positive tweets...")
top5_negative = list(topk_negative)[:5]
for word in top5_negative:
    freq1 = topk_negative[word]
    positive_lemmatized = df[df["label"] == 1].lemmatized
    freq2 = sum([lst.count(word) for lst in positive_lemmatized])
    print(f"{word}: ", pos_rate(freq1, freq2))
```

We can deduce that for this dataset, the most common words in positive tweets are strongly characteristic of positive tweets because the pos_rates are very high. 

On the other hand, the most common words in negative tweets are not characteristic of negative tweets because the pos_rates are much lower (except for "miss"). In fact, the words "get", "go", and "no" appear more frequently in positive tweets than negative tweets. 

This may seem like a surprising result. However it actually highlights how the overall sentiment of a tweet should not be decided simply by looking at the words the tweet contains, and therefore supports the need for a smarter classifier.
